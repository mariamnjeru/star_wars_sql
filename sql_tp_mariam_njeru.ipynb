```python
import requests
import sqlite3
import pandas as pd
```

```python
#data extraction for people
url = 'https://swapi.dev/api/people/'
all_people = []

while url:
    response = requests.get(url)
    data = response.json()
    results = data.get('results', [])
    print(results)
    all_people.extend(results)
   
```

```python
# cleaning of people data
df_people = pd.DataFrame(all_people)
columns_to_drop = ['films','created','edited','created','vehicles','species',]  #
df_people.drop(columns_to_drop, axis=1, inplace=True)
df_people

```

```python
#some more cleaning
def extract_id(url):
    if pd.isnull(url) or not isinstance(url, str):
        return url
    return int(url.strip('/').split('/')[-1])

df_people['planet_id'] = df_people['url'].apply(extract_id)

if 'homeworld' in df_people.columns:
    df_people['homeworld'] = df_people['homeworld'].apply(extract_id)

df_people
```

```python
#converting dtypes to strings
for column in df_people:
    if df_people[column].dtype == 'object':
        df_people[column] = df_people[column].astype(str)
```

```python
#same procedure for planets
url = 'https://swapi.dev/api/planets/'
all_planets = []

while url:
    response = requests.get(url)
    data = response.json()
    results = data.get('results', [])
    #print(results)
    all_planets.extend(results)
    url = data.get('next') 
 
```

```python
df_planets= pd.DataFrame(all_planets)
df_planets.head()
```

```python
df_planets = pd.DataFrame(all_planets)
columns_to_drop = ['films','created','edited']  #
df_planets.drop(columns_to_drop, axis=1, inplace=True)
df_planets.rename(columns={'name': 'planet_name'}, inplace = True)

def extract_id(url):
    if pd.isnull(url) or not isinstance(url, str):
        return url
    return int(url.strip('/').split('/')[-1])

df_planets['unique_planet_id'] = df_planets['url'].apply(extract_id)

df_planets.head()



```

```python
#cursor.execute('DROP TABLE IF EXISTS people')
for column in df_planets:
    if df_planets[column].dtype == 'object':
        df_planets[column] = df_planets[column].astype(str)
```

```python
# Step 4: SQLite Database Creation
with sqlite3.connect('star_wars.db') as conn:
    cursor = conn.cursor()
```

```python
cursor.execute(f"CREATE TABLE IF NOT EXISTS planets ({', '.join(df_planets.columns)})")
conn.commit()
df_planets.to_sql('planets', conn, if_exists='replace', index=False)
```

```python
#SQL SELECT
pd.read_sql(
    '''
        SELECT 
            people.name
        FROM 
            people
    
    ''', 
    conn
)
```

```python
#inner join to find out which person is from which planet
pd.read_sql(
    '''
        SELECT 
            people.name,
            planets.planet_name
        FROM 
            people
        INNER JOIN 
            planets ON people.planet_id = planets.unique_planet_id
        
    ''', 
    conn
)
```

```python
#left join to find the eye_color and planet_name :D
pd.read_sql(
    '''
        SELECT 
            people.name,
            people.eye_color,
            planets.planet_name
        FROM 
            people
        LEFT JOIN 
            planets ON people.planet_id = planets.unique_planet_id
        
    ''', 
    conn
)
```

```python
#Group by to find out the eyecolor frequency of people from different planets
pd.read_sql(
    '''
        SELECT 
            people.eye_color,
            COUNT (people.eye_color) as nb_eye_color,
            planets.planet_name
        FROM 
            people
        LEFT JOIN 
            planets ON people.planet_id = planets.unique_planet_id
        GROUP BY
        people.eye_color
        
    ''', 
    conn
)
```

```python
# having to find out the relation between skin_color and planet climate
pd.read_sql(
    '''
        SELECT 
            planet_name,
            people.skin_color,
            COUNT (people.skin_color) as nb_skin_color,
            planets.climate
        FROM 
            people
        LEFT JOIN 
            planets ON people.planet_id = planets.unique_planet_id
        GROUP BY
        people.skin_color
        HAVING
        nb_skin_color > 3
        
    ''', 
    conn
)
```

```python
#CTE
pd.read_sql(
    '''WITH skin_climate as (
        SELECT 
            planet_name,
            people.skin_color,
            COUNT (people.skin_color) as nb_skin_color,
            planets.climate
        FROM 
            people
        LEFT JOIN 
            planets ON people.planet_id = planets.unique_planet_id
        GROUP BY
        people.skin_color
        HAVING
        nb_skin_color > 3
    )

    SELECT 
    planet_name,
    skin_color
    FROM
    skin_climate
    WHERE
    climate = "arid"
    ''', 
    conn
)
```
